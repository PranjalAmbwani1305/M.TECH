```python
import pandas as pd
import numpy as np
```

```python
corpus = ['data science is one of the most important fields of science',
          'this is one of the best data science courses',
          'data scientists analyze data']
```

```python
words_set = set()
for doc in corpus:
    words = doc.split()
    words_set = words_set.union(set(words))
print('No of words in the corpus:',len(words_set))
print('The words in the corpus:\n',words_set)
```

```python
n_docs = len(corpus)
n_words_set = len(words_set)
words_list = list(words_set)  
df_tf = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=words_list)

for i in range(n_docs):
    words = corpus[i].split(' ')
    for w in words:
        # Use .loc for proper DataFrame indexing
        df_tf.loc[i, w] = df_tf.loc[i, w] + (i/len(words))

df_tf
```

```python
print("IDF of :")
idf = {}
for w in words_set:
    k = 0
    for i in range(n_docs):
        if w in corpus[i].split():
            k += 1
    idf[w] = np.log10(n_docs / k)
    print(f'{w:>15}: {idf[w]:>10}')
```

```python
n_docs = len(corpus)
n_words_set = len(words_set)
words_list = list(words_set)
df_tf = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=words_list)

for i in range(n_docs):
    words = corpus[i].split(' ')
    for w in words:
        df_tf.loc[i, w] = df_tf.loc[i, w] + (i/len(words))

df_tf_idf = pd.DataFrame(np.zeros_like(df_tf), columns=df_tf.columns)

for i in range(n_docs):
    words = corpus[i].split(' ')
    for w in words:
        df_tf_idf.loc[i, w] = df_tf.loc[i, w] * idf[w]

df_tf_idf
```
